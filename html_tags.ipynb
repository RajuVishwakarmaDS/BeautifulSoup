```python
from bs4 import BeautifulSoup
import requests

url="https://ritgoa.ac.in/electronics-telecommunication-engineering/"
r=requests.get(url)
content=r.content
soup=BeautifulSoup(content)

# get title tag
title=soup.find('title')
print(title.get_text())

# get head tag
head=soup.find('head')
print(head)

# get body tag
body=soup.find('body')
print(body.get_text())
```
